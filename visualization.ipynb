Why choose Martpotlib and Seaborn?


Both librarires are mong the most popular choices for data visualization  in python.Heres why: Matpotlib allows detailed customization of visual elements making it perfect for tailored plots.

Seaborn: Built on top of Matpotlib,Seaborn  simplifies visualization tasks by offering  high level functions to create aesthically pleasing statistical graphics.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

```

```python
#Line plot

x= np.linspace(0, 10, 100)
y= np.sin(x)
plt.show()
plt.plot(x, y)
plt.show()
```

```python
x= np.linspace(0, 10, 100)
y= np.sin(x)
plt.plot(x, y, label="sine wave")
plt.title("sine wave Line plot")
plt.xlabel("value of x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

```python
#bar chart
categories = ['A','B','C']
values = [50, 100, 75]
plt.bar(categories, values, color=['red', 'green', 'blue'])
plt.title("Bar chart example")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
#histogram
data = np.random.randn(1000)
plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value Bins")
plt.ylabel("Frequency")
plt.show()
```

```python

data = np.random.randn(1000)

n, bins, patches = plt.hist(data, bins=20, edgecolor='black')

colors = ['red', 'green', 'blue', 'orange', 'purple', 'pink', 'cyan', 'yellow', 'brown', 'gray']
for patch, color in zip(patches, colors * 2):
    patch.set_facecolor(color)

plt.title("Histogram Example")
plt.xlabel("Value Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
#histogram and KDE plot
data= np.random.randn((1000))
sns.histplot(data, kde=True, color='blue', edgecolor='black')

plt.title("Histogram with KDE Plot")
plt.xlabel("Value Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
#heatmap
matrix= np.random.rand(5, 5)
sns.heatmap(matrix, annot=True, cmap="viridis")
plt.title("Heatmap exmple")
plt.show()
```

```python
#pair plot
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.show()
```
